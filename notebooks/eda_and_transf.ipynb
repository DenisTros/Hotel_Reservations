V-CRAMER

```python
def cramers_V(var1,var2) :
    crosstab = np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
    stat = stats.chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
    return (stat/(obs*mini))

def get_cramers_df(data):
    rows = []
    for var1 in data:
        col = []
        for var2 in data:
            cramers = cramers_V(data[var1], data[var2]) # Cramer's V test
            col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
        rows.append(col)
    cramers_results = np.array(rows)
    return pd.DataFrame(cramers_results, columns = data.columns, index= data.columns)
```

## Outliers

```python
def tukey_outliers(df,column,extreme=False):
    q1, q3 = np.percentile(df[column],[25,75])
    iqr = q3 - q1
    constant = 1.5 if not extreme else 3
    return df[~((df[column]>(q3+constant*iqr)) | (df[column]<(q1-constant*iqr)))] 
```

```python
outlier_check_cols = ['lead_time','avg_price_per_room']

for columna in outlier_check_cols:
    outliers = round((1 - len(tukey_outliers(df_filtered,columna,extreme=False))/len(df_filtered))*100,2)
    print('Outliers en columna',columna,':',outliers,'%')
```

```python
df_clean = tukey_outliers(df_filtered,'lead_time',extreme=False)
df_clean = tukey_outliers(df_clean,'avg_price_per_room',extreme=False)
```

```python
df_clean.drop(columns=['booking_status','arrival_year','arrival_month','arrival_date','date'],inplace=True)
```

```python
sss = StratifiedShuffleSplit(n_splits =1, test_size =0.2 , random_state =42)
X_train , X_test , y_train , y_test = None , None , None , None

for train_index , test_index in sss.split(df_clean.drop(columns =['status']) ,df_clean['status']):
    train_df = df_clean.iloc[train_index]
    test_df = df_clean.iloc[test_index]
    X_train , y_train = train_df.drop(columns =['status']) , train_df.status
    X_test , y_test = test_df.drop(columns =['status']) , test_df.status
```
